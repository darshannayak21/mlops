```python
df = df[['date', 'venue', 'result', 'gf', 'ga', 'opponent', 
         'xg', 'xga', 'poss', 'sh', 'sot', 'pk', 'team', 'season']]
df.head()
```

## 1. Data Cleaning & Feature Engineering
Converting categories to numeric codes and creating rolling averages to avoid data leakage.


```python
import warnings
warnings.filterwarnings('ignore')
df['date'] = pd.to_datetime(df['date'])
df['venue_code'] = df['venue'].astype('category').cat.codes
df['opp_code'] = df['opponent'].astype('category').cat.codes
df['team_code'] = df['team'].astype('category').cat.codes
df['day_code'] = df['date'].dt.dayofweek
df['target'] = (df['result'] == 'W').astype(int)
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values('date')
    rolling_stats = group[cols].rolling(5, closed='left').mean()
    group[new_cols] = rolling_stats
    return group

cols = ['gf', 'ga', 'xg', 'xga', 'poss', 'sh', 'sot', 'pk']
new_cols = [f'{c}_rolling' for c in cols]

df = df.groupby('team', group_keys=False).apply(lambda x: rolling_averages(x, cols, new_cols))
df = df.dropna(subset=new_cols)
df = df.reset_index(drop=True)
df.head()
```

## 2. Visualization
Visualizing distributions and correlations.


```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='result', palette='viridis')
plt.title('Distribution of Match Results')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
corr = df[['target', 'venue_code', 'opp_code', 'day_code'] + new_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Heatmap')
plt.show()
```

## 3. Model Building
Splitting data temporally to test model generalizing to future matches.


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, classification_report

train = df[df['date'] < '2024-01-01']
test = df[df['date'] >= '2024-01-01']

predictors = ['venue_code', 'opp_code', 'team_code', 'day_code'] + new_cols

rf = RandomForestClassifier(n_estimators=100, min_samples_split=10, random_state=42)
rf.fit(train[predictors], train['target'])

preds = rf.predict(test[predictors])
print(f'Accuracy: {accuracy_score(test["target"], preds):.4f}')
print(f'Precision: {precision_score(test["target"], preds):.4f}')
print('\nClassification Report:')
print(classification_report(test['target'], preds))
```

```python
feature_imp = pd.Series(rf.feature_importances_, index=predictors).sort_values(ascending=False)
plt.figure(figsize=(10, 6))
feature_imp.plot(kind='bar', color='teal')
plt.title('Feature Importances in Random Forest Model')
plt.ylabel('Importance')
plt.show()
```

## 4. Export the Model
Saving the trained model as a `.pkl` file.


```python
import pickle
with open('model.pkl', 'wb') as f:
    pickle.dump(rf, f)
print('Model successfully exported to model.pkl')
```
